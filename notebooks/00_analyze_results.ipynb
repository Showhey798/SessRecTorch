```python
gru_sqn_recs = pd.read_csv("/home/inoue/work/program_016_inoue/master_programs/results/RC15/GRUSQNRec_rec_result.csv")
```

```python
gru_sqn_recs.rename(columns={"0":"sessionId", "1":"selected_item"})
```

```python
for i in [1, 5, 20, 3, 22]:
    print("--------------------------------------------------------")
    print("state: ", gru_sqn_recs.loc[i, "2":"11"].values)
    print("recommend: ", gru_sqn_recs.loc[i, "12":].values) 
    print("trueId: ", gru_sqn_recs.loc[i, "1"])
```

```python
import numpy as np
gru_sqn_recs.loc[:, "1":"11"].apply(
    lambda x: np.sum(x[0] == x[1:].values) > 0, axis=1
)
```

```python
select_item_not_in_hist_idx = gru_sqn_recs[gru_sqn_recs.loc[:, "1":"11"].apply(
    lambda x: np.sum(x[0] == x[1:].values) == 0, axis=1
)].index

select_item_not_in_hist_and_recommend_idx = gru_sqn_recs.loc[select_item_not_in_hist_idx, :][pd.concat([gru_sqn_recs.loc[select_item_not_in_hist_idx, "1"], gru_sqn_recs.loc[select_item_not_in_hist_idx, "12":]], axis=1).apply(
    lambda x: np.sum(x[0] == x[1:].values) != 0, axis=1
    )].index
```

```python
select_item_not_in_hist_and_recommend_idx.shape
```

```python
random_samples = np.random.choice(select_item_not_in_hist_and_recommend_idx, size=6)

for i in random_samples:
    print("--------------------------------------------------------")
    print("state: ", gru_sqn_recs.loc[i, "2":"11"].values)
    print("recommend: ", gru_sqn_recs.loc[i, "12":].values) 
    print("trueId: ", gru_sqn_recs.loc[i, "1"])
```
